```python
# アヤメのデータセットの読み込み
from sklearn import datasets

iris = datasets.load_iris()
print(iris)
```

```python
# 特徴量や分類の名前を確認
print("特徴量の名前=", iris.feature_names)
print("分類の名前=", iris.target_names)
print("分類の値=", iris.target)
```

```python
# データをデータフレームに入れる
import pandas as pd

df = pd.DataFrame(iris.data)
df.head()
```

```python
# 列名に特徴量を設定して、どんな品種なのかをtargetとして追加
df.columns = iris.feature_names
df["target"] = iris.target
df.head()
```

```python
# ヒストグラムで描画
import matplotlib.pyplot as plt

df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]

plt.figure(figsize=(5,5))
xx = "sepal width (cm)"
df0[xx].hist(color="b",alpha=0.5)
df1[xx].hist(color="r",alpha=0.5)
df2[xx].hist(color="g",alpha=0.5)
plt.xlabel(xx)
plt.show() # がくの幅から品種による違いがみられる
```

```python
# 2つの特徴量から散布図を作成する
import matplotlib.pyplot as plt

df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]

xx = "sepal width (cm)"
yy = "sepal length (cm)"
plt.figure(figsize=(5,5))
plt.scatter(df0[xx], df0[yy], color="b", alpha=0.5)
plt.scatter(df1[xx], df1[yy], color="r", alpha=0.5)
plt.scatter(df2[xx], df2[yy], color="g", alpha=0.5)
plt.xlabel(xx)
plt.ylabel(yy)
plt.grid()
plt.show()
```

```python
# 3つの特徴量から3D散布図を作成する
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

df0 = df[df["target"]==0]
df1 = df[df["target"]==1]
df2 = df[df["target"]==2]

xx = "sepal width (cm)"
yy = "sepal length (cm)"
zz = "petal length (cm)"

# 書籍から修正
fig = plt.figure(figsize=(5,5))
ax = Axes3D(fig)
fig.add_axes(ax)

ax.scatter(df0[xx], df0[yy], df0[zz], color="b", alpha=0.5)
ax.scatter(df1[xx], df1[yy], df1[zz], color="r", alpha=0.5)
ax.scatter(df2[xx], df2[yy], df2[zz], color="g", alpha=0.5)
ax.set_xlabel(xx)
ax.set_ylabel(yy)
ax.set_zlabel(zz)
ax.view_init(90,270)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# 散布図に分類の状態を描画する関数
def plot_boundary(model, X, Y, target, xlabel, ylabel):
    # 点と塗りのカラーマップ
    cmap_dots = ListedColormap([ "#1f77b4", "#ff7f0e", "#2ca02c"])
    cmap_fills = ListedColormap([ "#c6dcec", "#ffdec2", "#cae7ca"])

    plt.figure(figsize=(5, 5))
    # モデルがあれば、表示範囲の点をすべて予測して色を塗る
    if model:
        # 表示範囲を少し広げて分割し、調べる点（200x200）を用意する
        XX, YY = np.meshgrid(
            np.linspace(X.min()-1, X.max()+1, 200),
            np.linspace(Y.min()-1, Y.max()+1, 200))
        # 全ての点の値を、モデルで予測する
        pred = model.predict(np.c_[XX.ravel(), YY.ravel()]).reshape(XX.shape)
        #　予測結果の値（0〜2）の色（cmap_fills）で塗りつぶす
        plt.pcolormesh(XX, YY, pred, cmap=cmap_fills, shading="auto")
        #　境界を灰色で塗る
        plt.contour(XX, YY, pred, colors="gray")
    #　targetの値（0〜2）の色（cmap_dots）で点を描画する
    plt.scatter(X, Y, c=target, cmap=cmap_dots)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```
